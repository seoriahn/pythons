2. 이후 입력을 넣어준 값을 가진 헤더의 열 값을 나머지 열에서 빼주고 해당 지정된 열은 삭제
3. 열을 삭제하고 맨 오른족 열에 각 행의 평균값과 표준편차를 차례대로 입력
4. 맨 왼쪽 열에 입력된 값의 배수로 진행되는 time 열 추가

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
df
```

```python
df.columns.to_list()
```

```python
df=df.loc[:, ~df.columns[::-1].duplicated()[::-1]]
```

```python
col_name = input("column name: ")
```

```python
df = df.sub(df[col_name], axis=0)
df = df.drop(columns = [col_name])
```

```python
df
```

```python
temp = pd.DataFrame()
temp['mean'] = df.mean(axis=1)
temp['std'] = df.std(axis=1)

df['mean']=temp['mean']
df['std']=temp['std']
df
```
